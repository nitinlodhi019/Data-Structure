# Bubble Sort
Bubble sort is a sorting algorithm that starts from the first element of an array and compares it with the second element. If the first element is greater than the second, we swap them. It continues this process until the end of the array, with the largest elements “bubbling” to the top.

image.png



```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(n-1-i):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]

  return arr

arr = list(map(int,input().rstrip().split()))
print('unsorted list: ',arr)
result = bubble_sort(arr)
print('sorted list: ',result)
```

# Selection Sort

Selection sort takes the smallest element in an unsorted array and brings it to the front. You'll review each item (left to right) until you find the smallest one. The first item in the array is now sorted, while the rest of the array is unsorted.


image.png

```python
def selection_sort(arr):
  n = len(arr)

  for i in range(n-1):
    min_index = i
    for j in range(i+1, n):
      if arr[j] < arr[min_index]:
        min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr


arr = list(map(int,input().rstrip().split()))
print('unsorted list: ',arr)
result = selection_sort(arr)
print('sorted list: ',result)
```

# Insertion sort

Insertion sort in Python is a sorting algorithm that places elements in a list in order by comparing them. It's a simple algorithm that's good for small data sets.

**How it works**

The algorithm maintains two parts of the list: the sorted part and the unsorted part.

The algorithm takes an element from the unsorted part.

The algorithm compares the element to the elements in the sorted part.

The algorithm moves the element to the correct position in the sorted part.

The algorithm repeats the process for each element in the unsorted part.

image.png

```python
def insertion_sort(arr):
  n = len(arr)

  for i in range(1, n):
    for j in range(i, 0, -1):
      if arr[j] < arr[j-1]:
        arr[j], arr[j-1] = arr[j-1], arr[j]

  return arr

arr = list(map(int,input().rstrip().split()))
print('unsorted list: ',arr)
result = insertion_sort(arr)
print('sorted list: ',result)
```
